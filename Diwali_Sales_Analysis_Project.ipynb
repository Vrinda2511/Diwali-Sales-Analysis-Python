# Data Cleaning

removing duplicates,missing,outliers,convert datatypes,handling categorical,

```python
df.info()
```

```python
#to drop/remove unrelated columns which are null
df.drop(["Status","unnamed1"],axis=1,inplace=True)
```

```python
#to check for null values
pd.isnull(df).sum()
```

```python
 #to remove null values
df.dropna(inplace=True)
```

```python
pd.isnull(df).sum()  #check for null values-removed null
```

```python
#to change datatype of columns
df["Amount"]=df["Amount"].astype('int')
```

```python
df.info()
```

```python
df[["Age","Orders","Amount"]].describe()
```

```python
#to manage outliers
column_name=["Age","Orders","Amount"]
for i in column_name:
    print(i)
    plt.boxplot(df[i])
    plt.grid()
    plt.show()
```

```python
for i in column_name:    
    q75,q25=np.percentile(df[i],[75,25])

    IQR=q75-q25

    q0=q25-(IQR*1.5)
    q100=q75+(IQR*1.5)

    fdf=df[i][(df[i]<=q100) & (df[i]>=q0)]  #df created without outliers

    plt.boxplot(fdf)
    plt.grid()
    plt.show()
```

# Exploratory Data Analysis

Gender

```python
gn=sns.countplot(x="Gender",data=df)
for bars in gn.containers:
    gn.bar_label(bars)
```

```python
sales_gen=df.groupby(["Gender"],as_index=False)["Amount"].sum().sort_values(by="Amount",ascending=False)
gnamt=sns.barplot(x="Gender",y="Amount",data=sales_gen)
for bars in gnamt.containers:
    gnamt.bar_label(bars)
```

----From above graphs we can see that most buyers are females and have more purchasing power than men.

Age

```python
#On the basis of only age group as a whole
ag=sns.countplot(data=df,x= "Age Group")
for bars in ag.containers:
    ag.bar_label(bars)
```

```python
#On the basis of gender and Age
ag=sns.countplot(data=df,x= "Age Group",hue="Gender")
for bars in ag.containers:
    ag.bar_label(bars)
```

Marital Status

```python
ms = sns.countplot(data = df, x = 'Marital_Status')
sns.set(rc={'figure.figsize':(1,5)})
for bars in ms.containers:
    ms.bar_label(bars)
```

Occupation

```python
sns.set(rc={'figure.figsize':(20,5)})
occ = sns.countplot(data = df, x = 'Occupation')

for bars in occ.containers:
    occ.bar_label(bars)
```

Product Category

```python
sns.set(rc={'figure.figsize':(30,5)})
pc= sns.countplot(data = df, x = 'Product_Category')

for bars in pc.containers:
    pc.bar_label(bars)
```

From above graphs we can see that most of the sold products are from Food, Clothing and Electronics category

Married women age group 26-35 yrs from UP, Maharastra and Karnataka working in IT, Healthcare and Aviation are more likely to buy products from Food, Clothing and Electronics category
